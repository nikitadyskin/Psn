# Nova data frame com imformações extra sobre o ensino

```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
import seaborn as sns

df2 = pd.read_csv(r'data_V2_2010.csv')#juntei o df das aulas só com os valores de 2010 com dados de euducação de 2011 do pordata no excel
df2
```

# criação da matriz de correlação

```python
df_corr=df2.drop(['Ano','Densidade populacional'],axis=1)
df_corr= df_corr.iloc[1: , :]
corr_matrix=df_corr.corr()

corr_matrix


```

# visualização das correlações

```python
sns.heatmap(corr_matrix)
```

# visualização das correlações que têm o seu modulo entre 0.55 e 0.96 no excel 

```python
corr_matrix_abs = corr_matrix.abs()

for i in corr_matrix_abs.index:
     corr_matrix_abs[i]=corr_matrix_abs[(corr_matrix_abs[i]> 0.55) & (corr_matrix_abs[i] < 0.96)]
        
upper = corr_matrix_abs.where(np.triu(np.ones(corr_matrix_abs.shape), k=1).astype(np.bool))  

serie = upper.unstack().sort_values()



serie=serie.dropna()
corr_final= serie.to_frame()
corr_final.to_excel(r'corr.xls')
corr_final
```
