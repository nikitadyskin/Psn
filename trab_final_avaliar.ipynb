# Nova data frame com informações extra sobre o ensino

```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'data_V2_2010.csv')#juntei o df das aulas só com os valores de 2010 com dados de euducação de 2011 do pordata no excel
df
```

# criação da matriz de correlação

```python
df_corr=df.drop(['Ano','Densidade populacional'],axis=1)
df_corr= df_corr.iloc[1: , :]
corr_matrix=df_corr.corr()

corr_matrix
```

# visualização das correlações

```python
sns.heatmap(corr_matrix)
```

# visualização das correlações que têm o seu modulo entre 0.55 e 0.96 no excel 

```python
corr_matrix_abs = corr_matrix.abs()

for i in corr_matrix_abs.index:
     corr_matrix_abs[i]=corr_matrix_abs[(corr_matrix_abs[i]> 0.55) & (corr_matrix_abs[i] < 0.96)]
        
#para ficar com o triângulo de cima pois o de baixo tem valores repetidos        
upper = corr_matrix_abs.where(np.triu(np.ones(corr_matrix_abs.shape), k=1).astype(np.bool))  

serie = upper.unstack().sort_values()



serie=serie.dropna()
corr_final= serie.to_frame()
corr_final
corr_final.to_excel(r"corr.xlsx")
corr_final
```

# Criação de data frame para comparar institiçoes de ensino com o nível de instrução das pessoas  



```python
df2=df
#estabelecimentos

df2['estb_baixo'] = df['Estabelecimentos do ensino pré-escolar']+df['Estabelecimentos do 1.º ciclo do ensino básico']+ df['Estabelecimentos do 2.º ciclo do ensino básico']
df2['estab_medio'] = df['Estabelecimentos do 3.º ciclo do ensino básico']+df['Estabelecimentos do ensino secundário']
df2['estab_superior'] = df['Estabelecimentos do ensino superior']

#pessoas em cada nivel de escolaridade

df2['pessoas_baixo'] = df['(%)Básico 2º ciclo'] + df['(%) Básico 1º ciclo']
df2['pessoas_med'] = df['(%)Básico 3º ciclo'] + df['(%)Secundário']
df2['pessoas_superior'] = df['(%)Superior']



data = [df2['Municipios'],df2['estb_baixo'],df2['pessoas_baixo'],df2['estab_medio'],df2['pessoas_med'],df2['estab_superior'],df2['pessoas_superior']]

headers = ['Municipios','estb_baixo','pessoas_baixo','estab_medio','pessoas_med','estab_superior','pessoas_superior']

df3 = pd.concat(data, axis=1, keys=headers)

df3=df3.set_index('Municipios')

df3
```

# Criação de data frames para comparar institições de ensino com o nível de instrução das pessoas no litoral e no interior de Portugal continental 

```python
litoral = ['Lisboa', 'Porto', 'Coimbra', 'Setúbal', 'Braga']
interior = ['Aljustrel', 'Évora', 'Beja', 'Lamego', 'Bragança']


df_litoral=df3.loc[litoral]
df_interior=df3.loc[interior]
```

# Visualização dos dados do Litoral 

```python
def graficos(df_litoral,litoral):
    for i in range(len(litoral)):
        height = np.array(df_litoral)[i]
        bars = ('estb_baixo', 'pessoas_baixo', 'estab_medio', 'pessoas_med', 'estab_superior','pessoas_superior')
        x_pos = np.arange(len(bars))

        
        plt.rcParams['figure.figsize']=(10,6)
        
        # Create bars with different colors
        plt.bar(x_pos, height, color=['red', 'red', 'green', 'green', 'blue', 'blue'])

        # Create names on the x-axis
        plt.xticks(x_pos, bars)
        
        # Add title
        plt.title(litoral[i])
        
        
        # Show graph
        plt.show()        
```

```python
graficos(df_litoral,litoral) 
graficos(df_interior,interior)
```
