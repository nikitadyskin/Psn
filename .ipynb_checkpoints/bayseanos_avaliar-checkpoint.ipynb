```python

import pandas as pd
import numpy as np


#criar df e colunas, reorganizar  para os rings estarem em ordem crescente

df = pd.read_csv(r'C:\Users\nikit\OneDrive\Desktop\Psn\bayseanos\49541_data.csv',header=None)
df1=df

#mudar o index
df1.columns = df.iloc[0]

df1 = df.reindex(df1.index.drop(0)).reset_index(drop=True)

df1.columns.name = None
df1

#df1.head(10)

```

```python
#mudar os nomes para numeros
df1.replace(['Injetores','Pistons','Eletrico'],[1,2,3],inplace=True)
#df1=df1.replace(['Injetores','Pistons','Eletrico'],[1,2,3])
df1
```

```python
#calculo de probabilidades 

#converter dataframe toda para inteiros
df1=df1.astype(str).astype(int)

#probabilidade de doenca
P_D=[]

#probabilidade de fumo sabendo que doenca
F_D=[]

#probabilidade de falta de força sabendo que doenca
Ff_D=[]

#probabilidade de dificuldade em pegar sabendo que doenca
Dp_D=[]



#cria lista de doencas exitentes
Doencas=df1["Doenca"].unique().tolist()

Doencas.sort()

#conta os casos existentes
n_casos=np.count_nonzero(df1['Doenca'])

# preenche as listas com as respectivas probabilidades
for v in Doencas:
    P_D.append((np.count_nonzero(df1['Doenca']==v))/n_casos)
    F_D.append(np.count_nonzero((df1['Doenca']==v) & (df1["Fumo"] == 1))/(np.count_nonzero(df1['Doenca']==v)))
    Ff_D.append(np.count_nonzero((df1['Doenca']==v) & (df1["Falta de força"] == 1))/(np.count_nonzero(df1['Doenca']==v)))
    Dp_D.append(np.count_nonzero((df1['Doenca']==v) & (df1["Dificuldade em pegar"] == 1))/(np.count_nonzero(df1['Doenca']==v)))


    
    
#criar dataframe de probabilidades com as listas anteriores

Avaria=['Injetores','Pistons','Electrico']

Tp = pd.DataFrame(list(zip(Avaria,P_D, F_D,Ff_D,Dp_D)),
               columns =["Avaria","P problema", "P fumo |problema","P Falta de força|Problema)","P Dificuldade em pegar|Problema"])



Tp.set_index('Avaria', inplace=True)
Tp
```

```python
#camada 3 ->dificuldade em pegar

def prob_avaria_3(F,Ff,Dp):
    list_prob3=[]
    

    #versao 2 das listas de probabilidades
    F_D_2=F_D
    Ff_D_2=Ff_D
    Dp_D_2=Dp_D
    
    if F==0:
         F_D_2=[1-a for a in F_D]
    
    if Ff==0:
         Ff_D_2=[1-a for a in Ff_D]
    
    
    if Dp==0:
         Dp_D_2=[1-a for a in Dp_D]
    
    
    
    list_prob3 = [a*b*c*d for a,b,c,d in zip(P_D,F_D_2,Ff_D_2,Dp_D_2)]
    soma=sum(list_prob3)
    list_prob3 = [a/soma for a in list_prob3]
    
    
    return list_prob3
```

```python
#camada 2->falta de força

def prob_avaria_2(F,Ff):
    list_prob2=[]
    

    #versao 2 das listas de probabilidades
    F_D_2=F_D
    Ff_D_2=Ff_D
    
    if F==0:
         F_D_2=[1-a for a in F_D]
    
    if Ff==0:
         Ff_D_2=[1-a for a in Ff_D]
    
    
    list_prob2 = [a*b*c for a,b,c in zip(P_D,F_D_2,Ff_D_2)]
    soma=sum(list_prob2)
    list_prob2 = [a/soma for a in list_prob2]
    
    
    return list_prob2

```

```python
#camada 1->fumo

def prob_avaria_1(F):
    list_prob1=[]
    

    #versao 2 das listas de probabilidades
    F_D_2=F_D
 
    if F==0:
         F_D_2=[1-a for a in F_D]
    
    
    list_prob1 = [a*b for a,b in zip(P_D,F_D_2)]
    soma=sum(list_prob1)
    list_prob1 = [a/soma for a in list_prob1]
    
    
    return list_prob1

```
